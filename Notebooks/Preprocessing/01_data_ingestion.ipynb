```python
import pandas as pd
```

```python
file_path = '../Data/raw/'

date_columns = ['Date']

"Date","Open","Promo","StateHoliday","SchoolHoliday"
"Store","StoreType","Assortment","CompetitionDistance","CompetitionOpenSinceMonth","CompetitionOpenSinceYear","Promo2","Promo2SinceWeek","Promo2SinceYear","PromoInterval"


train_data_types_1 = {
    'Store': 'int16',
    'DayOfWeek': 'int8',
    'Customers': 'int16',
    'Open': 'bool',
    'Promo': 'int8',
    'SchoolHoliday': 'int8',
    'StateHoliday': 'category',
    'Sales': 'float32'
}
store_data_types = {
    'Store': 'int16',
    'StoreType': 'category',
    'Assortment': 'category',
    'CompetitionDistance': 'float32',
    'CompetitionOpenSinceMonth': 'float32',
    'CompetitionOpenSinceYear': 'float32',
    'Promo2': 'bool',
    'Promo2SinceWeek': 'float32',
    'Promo2SinceYear': 'float32',
    'PromoInterval': 'category'
}

test_data_types ={
    'Store': 'int16',
    'DayOfWeek': 'int8',
    'Customers': 'int16',
    'Open': 'float32',
    'Promo': 'int8',
    'SchoolHoliday': 'int8',
    'StateHoliday': 'category',
    'Sales': 'float32'
}
```

```python
df1 = pd.read_csv(
    file_path + "train.csv",
    parse_dates=date_columns,
    dtype=train_data_types_1,
    encoding="utf-8",
)
```

```python
df2 = pd.read_csv(
    file_path + "store.csv",
    dtype=store_data_types,
    
    encoding="utf-8",
)

training_data = pd.merge(df1, df2, on='Store', how='left')
training_data["Date_index"] = training_data["Date"]
training_data.set_index("Date_index", inplace=True)
training_data.head()
```

```python
testing_data = pd.read_csv(
    file_path + "test.csv",
    parse_dates=date_columns,
    dtype=test_data_types,
    encoding="utf-8",
)
testing_data = pd.merge(testing_data, df2, on='Store', how='left')
testing_data.head()
```

```python
testing_data["Open"] = testing_data["Open"].fillna(1)
testing_data["Open"] = testing_data["Open"].astype("bool")
testing_data.head()
```

```python
tr_col =training_data.columns
ts_col =testing_data.columns
```

```python
# ensure both datasets have the same columns
missing_cols = set(tr_col) - set(ts_col)
print("Columns in training data but not in testing data:", missing_cols)
```

```python
t2 = training_data[(training_data["Sales"] > training_data["Sales"].quantile(0.9995)) & (training_data.Open == 1)].sort_values("Sales", ascending=False)
t2[['Store', 'Promo', 'Promo2', 'PromoInterval', 'StoreType', 'StateHoliday', 'SchoolHoliday', 'Customers', 'Sales']].head(10)
```
